## Scatter Plot Matrix: 

Displays scatter plots for all pairs of variables in a dataset, helping to identify relationships and correlations

Grid of Axes:

The pair plot creates a grid of scatter plots, where each variable in the cont_df DataFrame is plotted against every other variable. This results in a matrix-like layout.

Scatter Plots:

Each scatter plot in the matrix shows the relationship between two variables. The x-axis represents one variable, while the y-axis represents another. Points in each scatter plot represent individual data points.

Diagonal:

The diagonal of the matrix usually contains histograms or kernel density estimates (KDE) that show the distribution of each variable. This helps to visualize the frequency of values for each variable

```python

# Create a scatter plot matrix
plt.figure(figsize=(20, 20))
sns.pairplot(df)
# Title for the figure
plt.suptitle('Scatter Plot Matrix'.upper(), y=1.02)  
plt.show()
```

## Heatmaps:
 Visualizes the correlation matrix or frequency of occurrences in a grid format, with color representing values

```python
# Create a DataFrame
df = pd.DataFrame(cont_df)

# Calculate the correlation matrix
correlation_matrix = cont_df.corr()
correlation_matrix
```

## Parallel Coordinates Plot:
 Represents multivariate data by drawing lines connecting individual observations across multiple axes, one for each variable.

Explaination of the graph

Variables (Columns):

Each vertical line in the plot represents a variable (column) from the cont_df DataFrame. The exact variables that appear depend on the content of cont_df.

Data Points:

Each line (or path) drawn across the vertical axes represents an individual data point from your DataFrame. Each point is connected across the variables, showing its values in relation to each variable.

Color Coding:

The lines are color-coded based on the categorical variable specified as col. This helps to distinguish between different groups or categories within the data.

```python
## Import library for creating paralell plot
from pandas.plotting import parallel_coordinates

##Creating the columns to be used
for i in cont_df.columns:
    col = i

# Create a parallel coordinates plot
plt.figure(figsize=(20, 20))
parallel_coordinates(cont_df, col, colormap='viridis')
plt.title('Parallel Coordinates Plot')
plt.xlabel('Variables')
plt.ylabel('Values')
plt.grid()
plt.show()
```

## 3D Scatter Plots: 
Allows visualization of relationships among three variables in a three-dimensional space.

Explaination of the plot

X-Axis (Unit Price):

This axis represents the unit price of a product or service. Each point's position along this axis indicates the price per unit.

Y-Axis (Quantity):

This axis represents the quantity of units sold or purchased. The position along this axis shows how many units are involved in the transaction.

Z-Axis (Total):

This axis represents the total revenue or total amount generated from the sale. It is usually calculated as the product of unit price and quantity

```python
# Create a 3D scatter plot
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot
ax.scatter(cont_df['Unit price'], cont_df['Quantity'],cont_df['Total'], c='g', marker='*')

# Labels and title
ax.set_xlabel('Unit Price')
ax.set_ylabel('Quantity')
ax.set_zlabel('Total')
ax.set_title('3D Scatter Plot'.upper())

plt.show()
```

## Bubble Charts: 
A variation of scatter plots where a third variable is represented by the size of the bubbles, adding another layer of information.

Explaination of the plot

X-Axis (Unit Price):

This axis represents the unit price of a product or service. Each point's position along this axis indicates how much the unit price is.

Y-Axis (Tax 5%):

This axis represents the amount of tax calculated at 5% on the unit price. The position along this axis shows how much tax is associated with the unit price.

Bubble Size (Gross Income):

The size of each bubble corresponds to the gross income associated with that specific unit price and tax amount. A larger bubble indicates a higher gross income, while a smaller bubble indicates a lower gross income.

```python
# Create a DataFrame
df = pd.DataFrame(cont_df)

# Create a bubble chart
plt.figure(figsize=(10, 6))
plt.scatter(cont_df['Unit price'], cont_df['Tax 5%'], s=cont_df['gross income'], alpha=0.5, c='blue', edgecolors='b', linewidth=0.7)

# Labels and title
plt.xlabel('Unit price')
plt.ylabel('Tax 5%')
plt.title('Bubble Chart'.upper())
plt.grid()

# Show the plot
plt.show()
```
